```python
import pandas as pd
import datetime
import re

data = pd.read_csv('C:/Users/SS/Desktop/AIFrienz S3_v1/target_v1.csv')

data['weekday'] = data['date'].apply(lambda x :datetime.datetime.strptime(x , "%Y-%m-%d").weekday())
kospi = pd.read_csv('C:/Users/SS/Desktop/AIFrienz S3_v1/코스피지수 내역.csv')

kospi['day_tmp'] = kospi['날짜'].apply(lambda x : re.findall("\d+", x ))
kospi['day'] =  kospi['day_tmp'].apply(lambda x : x[0]+'-'+x[1]+'-'+x[2])

kospi['종가'] = kospi['종가'].apply(lambda x : x.replace(',', ''))
kospi['오픈'] = kospi['오픈'].apply(lambda x : x.replace(',', ''))
kospi['고가'] = kospi['고가'].apply(lambda x : x.replace(',', ''))
kospi['저가'] = kospi['저가'].apply(lambda x : x.replace(',', ''))

kospi['종가'] = kospi['종가'].astype(float)
kospi['오픈'] = kospi['오픈'].astype(float)
kospi['고가'] = kospi['고가'].astype(float)
kospi['저가'] = kospi['저가'].astype(float)

kospi['변동 %'] = kospi['변동 %'].apply(lambda x : x.replace('%', ''))
kospi['변동 %'] = kospi['변동 %'].astype(float)

kospi['거래량'] = kospi['거래량'].apply(lambda x : x.replace('M', '') if x[-1] == 'M' else x.replace('B','0'))
kospi.loc[151 , '거래량'] =  '1010'
kospi['거래량'] = kospi['거래량'].astype(float)
kospi.drop(['날짜','day_tmp'], axis =1, inplace=True)
kospi.columns = ['CLOSE','OPEN','HIGH','LOW','VOLUME(Million)','Change_Rate(%)','date']

integ = pd.merge(data, kospi, how ='left')

integ.fillna(0,inplace=True) # 코스피 지수 정보가 없는 날의 값들은 0으로 처리하였음
integ
```
